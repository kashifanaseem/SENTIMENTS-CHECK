**5) Date of Tweets:** To check the dates where tweets had been done the most.

```python
dfD=df[["user_name","date"]]       #dataframe with two columns only user's name and the date of tweet
```

```python
print(dfD["date"].dtypes)         #type of date column
```

```python
df_users = pd.DataFrame()
df_users['date'] = pd.to_datetime(dfD['date']).dt.date             #changed the type in date time and again in object to remove the time from the date
df_users['user_name'] = dfD["user_name"]
```

```python
print(df_users["date"].dtypes)
```

```python
df_users[['user_name','date']].head(2)                      
#now how this is looking like without time, adding time will creat the dataframe much clustered so to make it more feasiable removed the time
```

```python
df_tweets = df_users["date"].value_counts().rename_axis('dates').reset_index(name='tweets')      #counting the tweets on every date
df_tweets.head(3)
```

```python
df_dashbord=df_tweets.sort_values(by="dates")
df_dashbord.head(3)                                          #sorted the data by date from 24/7 onwards
```

```python
import pandas as pd
from bokeh.plotting import figure, output_file, show,output_notebook
output_notebook()
```

```python
def make_dashboard(DATE, TWEETS, title, file_name):
    output_file(file_name)
    p = figure(title=title, x_axis_label='Date', y_axis_label='Tweets')
    p.line(DATE.squeeze(), TWEETS.squeeze(), color="firebrick", line_width=4, legend="tweets")
    show(p)                                                                          #making dashboard
```

```python
DATE=df_dashbord[["dates"]]                     
```

```python
TWEETS=df_dashbord[["tweets"]]
```

```python
title="Tweets_Dashboard"
```

```python
file_name="tweets.html"
```

```python
make_dashboard(DATE=DATE, TWEETS=TWEETS, title=title, file_name=file_name)
```

Most of the tweets were tweeted in the beginning, later it become constant and few times went lesser and sometime came a bit more, it is predictable from the trend that the sentiments on twitter were much higher in july the way people were facing the situation, but later they became used to of it or the situations are better for them so tweeted much lesser and the number of tweets everyday became constant.
